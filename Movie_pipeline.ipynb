```python
import os 
import requests
import pandas as pd 
from datetime import date
```

```python
def download(url,dest_folder):
    
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)
    
    file_name = url.split('/')[-1]
    file_path = os.path.join(dest_folder,file_name)
    
    r = requests.get(url)
    
    if r.ok:
        print(f'Saving file to {file_path}')
        with open(file_path,'wb') as f:
            f.write(r.content)
        f.close()
        
    else: print(f'Failed with code : {r.status_code}')
```

```python
today = date.today().strftime('%Y-%m-%d')
dest_folder = f'raw/day={today}'
url = 'https://raw.githubusercontent.com/RiccardoRubini93/Academy/main/movie_metadata_raw.csv'

download(url,dest_folder)
```
